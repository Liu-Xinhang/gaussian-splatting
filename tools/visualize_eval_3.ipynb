```python
from matplotlib import pyplot as plt
import numpy as np
from pathlib import Path
```

```python
root_path = Path("../track")
categories = root_path.glob("*")
for category in categories:
	degree_path = category / "200_degree.npy"
	cm_path = category / "200_cm.npy"
	degree = np.load(degree_path)
	plt.plot(degree, label=category.name.split("_")[-1].rstrip("-2"))
	plt.legend()
```
