```python
from matplotlib import pyplot as plt
import numpy as np
from pathlib import Path
import time
import ipywidgets as widgets
from IPython.display import display, clear_output

def load_data(total, idx):
	res0 = []
	res_end = []
	for item_path in total:
		res0.append(np.load(item_path)[0])
		res_end.append(np.load(item_path)[idx])
	return res0, res_end

def plot_image(total, idx, title=""):
	res0, res_end = load_data(total, idx)
	plt.plot(res0, label="begin")
	plt.plot(res_end, label="end")
	plt.legend()
	plt.title(title)
	plt.show()
```

```python
root_dir = Path("../debug_YCBV/debug_data_000002_000050")
total_add = sorted(root_dir.glob("*add.npy"))
total_adds = sorted(root_dir.glob("*adds.npy"))
total_cm = sorted(root_dir.glob("*cm.npy"))
total_degree = sorted(root_dir.glob("*degree.npy"))
```

```python
def show(total, title):
    def update_plot(change):
        clear_output(wait=True)
        plot_image(total, idx=change.new, title=title)
        display(slider)
    slider = widgets.IntSlider(value=1, min=1, max=41, step=1, description="Index from end:")
    # Set the function to be triggered on value change
    slider.observe(update_plot, names="value")

    # Initial display
    display(slider)
    plot_image(total, idx=slider.value, title=title)
```

```python
show(total_add, "add")
```

```python
show(total_cm, "cm")
```

```python
show(total_degree, "degree")
```
